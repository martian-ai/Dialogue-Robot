```python
import torch
import numpy as np
import torch.nn as nn
X = torch.rand(4, 2, 3)
linear = nn.Linear(3, 2)
```

```python
"""矩阵加法"""
# Z=W+X
W = torch.randn(4, 2, 3)
Z11 = W + X
print(Z11)
Z12 = W.add(X)
print(Z12)
```

```python
# Z = WX -线性变换
Z2 = linear(X)
print('m.weight.shape:', linear.weight.shape) # 转置
print('m.bias.shape:', linear.bias.shape)
print('output.shape: ', Z2.shape)
print(Z2)
print(Z2.size())
```

```python
# Z = W⊗X -矩阵外积 或 矩阵乘法[m,n]⊗[n,p]==[m,p]
mat1 = torch.randn(3, 4)
mat2 = torch.randn(4, 3)
print(torch.mm(mat1, mat2))
```

```python
Z32 = np.outer(X, W2)
print(Z32)
```

```python
Z33 = np.dot(X, W2)
print(Z33)
```

```python
Z34 = torch.dot()  # 仅能用于向量运算[1,n]⊗[n,1]
print(Z34)
# Z = W⊙X -矩阵元素积:即对应位置元素相乘[m,n]⊙[m,n]==[m,n]
```

```python
Z41 = torch.mul(X, W)
print(Z41)
```

```python
Z43 = W * X  # 若W和X是torch.tensor，同样是元素积
print(Z43)
```

```python
Z42 = np.multiply(X, W)
print(Z42)
```
