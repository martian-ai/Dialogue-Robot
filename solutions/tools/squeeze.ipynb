```python
import torch
x = torch.Tensor([1, 2, 3, 4])  # torch.Tensor是默认的tensor类型（torch.FlaotTensor）的简称。
print(x)  # tensor([1., 2., 3., 4.])
print(x.size())  # torch.Size([4])
print(x.dim())  # 1
print(x.numpy())  # [1. 2. 3. 4.]
```

# Unsqueeze
+ 不改变原始数据

```python
print(torch.unsqueeze(x, 0))  
print(x)
print(torch.unsqueeze(x, 1)) 
print(x)
print(torch.unsqueeze(x, -1)) 
print(x)
print(torch.unsqueeze(x, -2)) 
print(x)
```

# Unsqueeze_
+ 改变原始数据

```python
print(x.unsqueeze_(0))  
print(x)
```

# Squeeze
+ 不改变原始数据

```python
print(x)
print(x.squeeze(-1))
print(x)
print(x.squeeze(-2))
```

# Squeeze_
+ 改变原始数据

```python
print(x)
print(x.squeeze_(-1))
print(x)
print(x.squeeze(-1))
```

```python
print(x)
print(x.squeeze_(-2))
```
