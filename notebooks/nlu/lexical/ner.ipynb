```python
### Solution
+ 基于规则和词典
    + 基于规则的方法多采用语言学专家手工构造规则模板
    + 选用特征包括统计信息、标点符号、关键字、指示词和方向词、位置词(如尾字)、中心词等方法，以模式和字符串相匹配为主要手段，这类系统大多依赖于知识库和词典的建立
+ 基于统计的方法(特征工程）
    + HMM，ME， CRF，SVM, NN
    + 最大熵模型有较好的通用性，主要缺点是训练时间复杂性非常高
    + 条件随机场特征灵活、全局最优的标注框架，但同时存在收敛速度慢、训练时间长的问题
    + 隐马尔可夫模型在训练和识别时的速度要快一些，Viterbi算法求解命名实体类别序列的效率较高

+ 混合方法
自然语言处理并不完全是一个随机过程,单独使用基于统计的方法使状态搜索空间非常庞大，必须借助规则知识提前进行过滤修剪处理

+ 基于神经网路的序列标注
    + Lattice-LSTM CRF 
    + IDCNN CRF 
    + DBN CRF 
```

### Metric
+ strict/type/partial/overlap/
+ overlap
    + ture 和 false 的ner 有交叉, 但是类型不同
    + 如  "市长/江大桥" "市/长江大桥" 中， "江大桥" 和 “长江大桥" 的类型分别是 人名 和 地名
+ type
    + ture 和 false 的ner 有交叉, 但是类型不同
+ partal
    + ture 和 false 的ner 类型相同，且两者的边界存在包含关系
+ strict 严格正确
    + ture 和 false 的ner 类型和边界都正确
