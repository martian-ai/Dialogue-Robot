```python
from collections import Counter
sort_dict = Counter(text)
# print(sort_dict)
word_dict = { key:idx for idx, key in enumerate(sort_dict.keys())}
# print(word_dict)
```

```python
# Pointer mutual information
import numpy as np
co_occurrence_matrix = np.zeros((len(word_dict), len(word_dict)))
print(co_occurrence_matrix.shape)
```

```python
for tmp_idx, tmp_char in enumerate(text):
    if tmp_idx != len(text)-1:
        co_occurrence_matrix[word_dict.get(text[tmp_idx])][word_dict.get(text[tmp_idx+1])]+=1

np.set_printoptions(threshold=np.inf)
```

## Pointwise Mutual Information
$ PMI(x,y) = log \frac{p(x, y)}{p(x) * p(y)}$

```python
def PMI(a, b):
    p_ab = co_occurrence_matrix[word_dict.get(a)][word_dict.get(b)]
    p_b = sort_dict.get(b)
    p_a = sort_dict.get(a)
    # print(p_ab)
    # print(p_a)
    # print(p_b)
    
    if p_ab == 0:
        return 0
    else:
        return np.log(p_ab/(p_a * p_b)) 

print(PMI('一', '笑'))
print(PMI('三', '千'))
print(PMI('姊', '妹'))
```

## Mutual Information
$ MI(X,Y) = \sum_{x\in X} \sum_{y\in Y} p(x,y) log \frac{p(x, y)}{p(x) * p(y)}$

```python
def MI(a, b):
    all_results = []
    for tmp_a, tmp_b in zip(a, b):
        p_ab = co_occurrence_matrix[word_dict.get(tmp_a)][word_dict.get(tmp_b)]
        all_results.append(p_ab * PMI(tmp_a, tmp_b))
    
    return np.sum(all_results)
a = '峨嵋山下少人行'
b = '旌旗无光人来人往'
print('MI')
print(MI(list(a), list(b)))
```
