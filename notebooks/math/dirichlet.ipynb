### Dirichlet Function
$f(x)=\left\{\begin{array}{ll}1, & \text { x为有理数 } \\ 0, & \text { x为无理数 }\end{array}\right.$

### Dirichlet Distribution
+ $f(x)=\frac{1}{\mathrm{~B}(\boldsymbol{\alpha})} \prod_{i=1}^{K} x_{i}^{\alpha_{i}-1}$, 其中 $\mathrm{B}(\boldsymbol{\alpha})=\frac{\prod_{i=1}^{K} \Gamma\left(\alpha_{i}\right)}{\Gamma\left(\sum_{i=1}^{K} \alpha_{i}\right)}$， $\sum_{i=1}^{K} x_{i} \leq 1$

```python
from scipy.stats import dirichlet
import numpy as np 
quantiles = np.array([0.18, 0.22, 0.6])  # specify quantiles
alpha = np.array([1, 1, 12])  # specify concentration parameters
print(dirichlet.pdf(quantiles, alpha))
print(dirichlet.mean(alpha))  # get the mean of the distribution
print(dirichlet.var(alpha)) # get variance
print(dirichlet.entropy(alpha))  # calculate the differential entropy
```
